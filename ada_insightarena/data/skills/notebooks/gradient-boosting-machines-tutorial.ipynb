* **Before examining the Gradient Boosting algorithm, we should examine the gradient descent and boosting methods, which are the basis of Gradient Boosting.**

### What is Gradient Descent?

* Graident descent is an optimization method.
* It is used to find parameters that can minimize the Cost(error) function.
* Its purpose is to find the parameter values that can make any differentiable function a minimum.
* Gradient descent steps are as follows:
* It takes the partial derivative of the related function according to the related parameter.
* Obtained derivative (gradient) value gives the maximum increase direction of the related function.
* It updates the parameter by going against the direction of this maximum increase with a certain intensity.
* The landing intensity is decided by the learning rate.
* If the learning rate is high, the train time may be shorter, but we may not find the optimum result.
* If the learning rate is small, the train time will be longer, but we will get more accurate results.
* The new value of the parameter is created by multiplying the gradient value with the learning rate. and it ensures that the error is reduced in each iteration.
* In summary; It iteratively updates the parameter values in the steepest descent direction, which is defined as the negative of the gradient, and finds the parameters that can give the minimum values of the relevant function.

### What is Boosting?

* Boosting methods can be considered as an optimization method. (Breiman,2001).
* Boosting is a tree based method that tries to increase model performance based on residual optimization. Therefore, it is an adaptive method.
* It is based on the idea that weak classifiers come together to form a stronger classifier.
* Unlike the bagging method, the trees in the boosting method are dependent on each other because they progress over the residuals.
* A series of models in the form of a single predictive model are built on errors/residues.These models are created additively.
* AdaBoost was the first really successful boosting algorithm developed by Freund ve Schapire in 1999.
* In Adaboost (Adatptive Boosting), the data is first trained with a weak learner. After the training, the wrong predictions are retrained by increasing the weights.

## What is Gradient Boosting Machines?

* It is a tree-based ensemble learning algorithm formed by the combination of gradient descent and boosting methods.A series of models in the form of a single predictive model are built on errors/residues.These models are created additively.

* In the function below, f0(x) or F0(X) is called base learner.Also, F0(X) represents the predictions of the first model.

* Firstly, the prediction model is built on this base learner and as a result of this model, prediction values are obtained. The residuals / errors are obtained by taking the difference between the actual y(dependent variable) values and the estimated y_head values. 

* A new model is built with these residuals and independent variables, as if the residuals were the dependent variable.

* The prediction of the residual model is first multiplied by the learning rate and then added to the prediction of the previous model to obtain new y_head values.

*  Again by subtracting the new y_head values from the actual  y values, the residuals are obtained and the model is built on the residuals.After the predictions of the model built with these residuals are multiplied by the learning rate, they are added to the predictions of the previous model.
* This process repeats as many as the number of iterations determined
* Finally,the latest y_head value is obtained by summing the base learner prediction and the prediction values of all models built on residuals. fo(x) first prediction value and  All Δ(x) values represent the prediction values formed by modeling the residual.

* LR =Learning Rate
* y_head = f0(x) + Δ1 * (LR) + Δ2 * (LR) + Δ3 * (LR) + ........ + Δm * (LR)

* The logic of the algorithm and iterations will be better understood in the above numerical example. The sqfeet and rent values of a house are given. Our aim is to estimate the rent value correctly when the sqfeet value is given.

* F0 is the predicion of first decision tree model. F0 = first y_head
* Then the estimated y_head values are subtracted from the actual y values to find the residuals. ( y - F0 = Residuals )
* A new model is established with the residuals obtained and the prediction results of this model are shown as Δ1. After Δ1 is multiplied by the learning rate(LR), it is added to F0 and the F1 value is obtained. F1= new y_head

* F1 = F0 + Δ1 * (LR)

* And subtract the F1 values from the actual y values to obtain the new residuals ( y - F1 = new Residuals ). With these residuals, a new model is established and predictions called  Δ2 are obtained. After Δ2 is multiplied by the learning rate(LR), it is added to F1 and the F2 value is obtained.
* F2 = F1 + Δ2 * (LR)

* F(M) = F(M-1) + Δ(M) * (LR)
* This process repeats as many as the number of iterations determined with the help of gradient descent.
*  In this process, All Δ(X) values represent the prediction values formed by modeling the residuals, and F(X) values represent the new y_head values formed by adding these Δ(X)'s.

Gradient Boosting Machines has some important hyperparameters.They are divided into 3 among themselves.
1. **Tree-Specific Parameters** : These affect each individual tree in the model.
1. **Boosting Parameters**: These affect the boosting operation in the model.
1. **Miscellaneous Parameters**: Other parameters for overall functioning.

* **Boosting Parameters** : These affect the boosting operation in the model.
1. learning_rate : 

* Learning rate shrinks the contribution of each tree by learning_rate. There is a trade-off between learning_rate and n_estimators. If the learning rate is small, the n_estimators should be large.(vice-versa)

* The prediction obtained in each tree is updated by multiplying by the learning rate.This process is done with the help of gradient descent.With the partnership of learning rate and gradient descent, the error function is minimized and the most accurate y value is tried to be estimated.

* If the learning rate is high, the train time may be shorter, but we may not find the optimum result.If the learning rate is small, the train time will be longer, but we will get more accurate results.


2. n_estimators : 
* It is the number of trees to boost. In other words, it is the number of iterations. Too many iterations can cause overfitting. Therefore, the ideal number of iterations should be found with hyperparameter tuning.

3. Subsample: 
* Observation rate to be selected for each tree. The selection is made by random sampling.

4. Criterion: 
* The function to measure the quality of a split. Supported criteria are ‘friedman_mse’ for the mean squared error with improvement score by Friedman, ‘squared_error’ for mean squared error, and ‘mae’ for the mean absolute error. The default value of ‘friedman_mse’ is generally the best as it can provide a better approximation in some cases.

* **Tree-Specific Parameters**
1.  min_samples_split:

* Minimum number of samples a node must possess before splitting.
for example; when min_samples_split=4 ,a node must have at least 4 sample to be split.

2. max_leaf_nodes: 
* Maximum number of leaf nodes a decision tree can have.
3. max_depth:
* The maximum depth of the tree. If None, then nodes are expanded until all leaves are pure or until all leaves contain less than min_samples_split samples.If max_depth is too much, the tree may branch too much and overfitting may occur.
4. min_samples_leaf:
* Minimum number of samples a leaf node must possess.

5. max_features: 
* It is the number of features to be evaluated for splitting. options=‘auto’, ‘sqrt’, ‘log2’,specific value. Also we can find the best number of max_features by giving some values or ranges in hyperparameter optimization.

* All these hyperparameters help pruning and try to avoid overfitting problem.Hyperparameter optimization should be done to find the best combination of hyperparameters.

* **Miscellaneous Parameters**:

1. **n_iter_no_change:** 

* n_iter_no_change is used to decide if early stopping will be used to terminate training when validation score is not improving. By default it is set to None to disable early stopping. If set to a number, it will set aside validation_fraction size of the training data as validation and terminate training when validation score is not improving in all of the previous n_iter_no_change numbers of iterations. The split is stratified. Values must be in the range [1, inf).
  
  
2. **validation_fraction:**
* The proportion of training data to set aside as validation set for early stopping. Values must be in the range (0.0, 1.0). Only used if n_iter_no_change is set to an integer.

```python
# Necessary libraries
import warnings
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import GridSearchCV, cross_validate, validation_curve
from matplotlib import pyplot as plt
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import classification_report, roc_auc_score,accuracy_score
warnings.simplefilter(action='ignore', category=Warning)
from sklearn.model_selection import learning_curve
```

```python
df=pd.read_csv("../input/ionosphere-data/ionosphere.csv",sep=",")
df.dropna(inplace=True)
df.head()
```

* **Dataset Information**
* This radar data was collected by a system in Goose Bay, Labrador. This system consists of a phased array of 16 high-frequency antennas with a total transmitted power on the order of 6.4 kilowatts. See the paper for more details. The targets were free electrons in the ionosphere. "Good" radar returns are those showing evidence of some type of structure in the ionosphere. "Bad" returns are those that do not; their signals pass through the ionosphere.

* Received signals were processed using an autocorrelation function whose arguments are the time of a pulse and the pulse number. There were 17 pulse numbers for the Goose Bay system. Instances in this databse are described by 2 attributes per pulse number, corresponding to the complex values returned by the function resulting from the complex electromagnetic signal.

* All 34 are continuous. The 35th attribute is either "good" or "bad" according to the definition summarized above. This is a binary classification task.

```python
# converting from string to numeric
df["class"]=df["class"].map({"g":1,"b":0})
```

```python
y = df["class"] # dependent(target) variable
X = df.drop(["class"], axis=1) # independent variables
```

```python
# create GBM model
gbm_model = GradientBoostingClassifier(n_iter_no_change=5,
                                       validation_fraction=0.20,
                                       random_state=17).fit(X,y)
```

```python
cv_results = cross_validate(gbm_model, X, y, cv=10, scoring=["f1"],
                            return_train_score=True)

print("train f1 score:",cv_results['train_f1'].mean())
print("test f1 score:",cv_results['test_f1'].mean())

# using f1 score for evaluation metric because we have imbalanced dataset
#The gbm classifier model we created gives very good results in both train and test set
```

```python
train_sizes, train_scores, test_scores = learning_curve(gbm_model, X, y, cv=10,
                                                        scoring='f1', n_jobs=-1, 
                                                        train_sizes=np.linspace(0.01, 1.0, 100))

train_mean = np.mean(train_scores, axis=1)
validation_mean = np.mean(test_scores, axis=1)

plt.style.use('seaborn')
plt.plot(train_sizes, train_mean, label = 'Training error')
plt.plot(train_sizes, validation_mean, label = 'Validation error')
plt.ylabel('Accuracy', fontsize = 14)
plt.xlabel('Training set size', fontsize = 14)
plt.title('Learning curve for GBM', fontsize = 18, y = 1.03)
plt.legend();
```

```python
def val_curve_params(model, X, y, param_name, param_range, scoring="roc_auc", cv=10):
    train_score, test_score = validation_curve(
        model, X=X, y=y, param_name=param_name, param_range=param_range, scoring=scoring, cv=cv)

    mean_train_score = np.mean(train_score, axis=1)
    mean_test_score = np.mean(test_score, axis=1)

    plt.plot(param_range, mean_train_score,
             label="Training Score", color='b')

    plt.plot(param_range, mean_test_score,
             label="Validation Score", color='g')

    plt.title(f"Validation Curve for {type(model).__name__}")
    plt.xlabel(f"Number of {param_name}")
    plt.ylabel(f"{scoring}")
    plt.tight_layout()
    plt.legend(loc='best')
    plt.show(block=True)
```

```python
val_curve_params(gbm_model, X, y, "n_estimators", range(1,101,10), scoring="f1")
```

```python
val_curve_params(gbm_model, X, y, "learning_rate", np.arange(0.03,0.1,0.01), scoring="f1")
```

```python
val_curve_params(gbm_model, X, y, "max_depth", range(1,6), scoring="f1")
```

```python
val_curve_params(gbm_model, X, y, "min_samples_leaf", range(1,41,5), scoring="f1")
```

```python
gbm_params = {"learning_rate": [0.07,0.08],
              "max_depth": [1,2,3],
              "n_estimators": [10,20,30,40,50],
              "subsample": [0.5,0.6],
              "min_samples_split": range(12,16),
              "min_samples_leaf" : range(14,19),
              "max_features":[7,10,13]
             }

gbm_best_grid = GridSearchCV(gbm_model, gbm_params, cv=5, n_jobs=-1, verbose=True).fit(X, y)
```

```python
gbm_best_grid.best_params_
```

```python
gbm_best_grid.best_score_
```

```python
gbm_final = gbm_model.set_params(**gbm_best_grid.best_params_,
                                 random_state=17, ).fit(X, y)
# create final model with best gridsearch hyperparameters
```

```python
cv_results_final = cross_validate(gbm_final, X, y, cv=10, 
                            scoring=["f1"],
                            return_train_score=True)
print("train f1 score:",cv_results_final['train_f1'].mean())
print("test f1 score:",cv_results_final['test_f1'].mean())
```

```python
train_sizes_final, train_scores_final, test_scores_final = learning_curve(gbm_final, X, y,
                                                                          cv=10,
                                                        scoring='f1',n_jobs=-1, 
                                                        train_sizes=np.linspace(0.01, 1.0, 100))

train_mean_final = np.mean(train_scores_final, axis=1)
validation_mean_final = np.mean(test_scores_final, axis=1)

plt.style.use('seaborn')
plt.plot(train_sizes_final, train_mean, label = 'Training error')
plt.plot(train_sizes_final, validation_mean, label = 'Validation error')
plt.ylabel('Accuracy', fontsize = 14)
plt.xlabel('Training set size', fontsize = 14)
plt.title('Learning curve of Final GBM', fontsize = 18, y = 1.03)
plt.legend();
```

* We analyzed the complexity of the model specific to the hyperparameters using a validation curve and observed where the f1 score increases and decreases according to the hyperparameters. As a result of these observations, we determined the intervals where the hyperparameters would give the best results and we performed the hyperparameter optimization with gridsearchcv.
* As a result of gridsearchcv, train score decreased slightly, while test score increased. This made our GBM model more resistant to overfitting.

```python
def plot_importance(model, features, num=len(X), save=False):
    feature_imp = pd.DataFrame({'Value': model.feature_importances_, 'Feature': features.columns})
    plt.figure(figsize=(8, 8))
    sns.set(font_scale=1)
    sns.barplot(x="Value", y="Feature", data=feature_imp.sort_values(by="Value",
                                                                     ascending=False)[0:num])
    plt.title('Features')
    plt.tight_layout()
    plt.show()
    if save:
        plt.savefig('importances.png')

plot_importance(gbm_final, X)
```

* When the feature importance graph is examined, the most important variables are; It is seen that a05,a27,a07,a03.Variables a32 and a02 have almost no effect on the model.
