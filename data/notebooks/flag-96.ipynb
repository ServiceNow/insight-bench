### **Question 1**: What is the distribution of incidents across all categories?

#### Plot number of incidents by category


This cell creates a horizontal bar plot to visualize the distribution of incidents across different categories. It annotates the plot with the count of incidents in each category. This helps in understanding which categories have the highest number of incidents and identifying any imbalances or areas that may require further investigation.


```python

plot = df.groupby("category").size().plot(kind="barh", color=sns.palettes.mpl_palette("Dark2"))

fig = plt.gcf()


for i in plot.patches:
    # Get X and Y placement of label from rectangle
    x_value = i.get_width()
    y_value = i.get_y() + i.get_height() / 2

    # Use X value as label and format number with one decimal place
    label = "{:.1f}".format(x_value)

    # Create annotation
    plt.annotate(
        label,                      
        (x_value, y_value),         
        xytext=(-10, 0),            
        textcoords="offset points", 
        ha='right',                 
        va='center'                 
    )

# Set plot title
plt.title('Incidents Distribution by Category')

# Set x-axis label
plt.xlabel('Category')

# Set y-axis label
plt.ylabel('Number of Incidents')

# Display the figure
plt.show()
```

### **Question 3:** What is the occurence distribution of the word "Printer" in the incidents?

### **Question 5**: Is there a pattern or trend over time in the distribution of incidents across categories?

#### Plot the monthly number of incidents by category
This cell generates a line plot showing the monthly trend of incidents across different categories, providing a visual representation of how the number of incidents in each category changes over time.

```python
# Convert 'opened_at' to datetime
df['opened_at'] = pd.to_datetime(df['opened_at'])

# Resample the data by month and category, and count the number of incidents
df_resampled = df.groupby([pd.Grouper(key='opened_at', freq='M'), 'category']).size().unstack()

# Plot the resampled data
plot = df_resampled.plot(kind='line')

# Set plot title
plt.title('Incidents Over Time by Category')

# Set x-axis label
plt.xlabel('Time')

# Set y-axis label
plt.ylabel('Number of Incidents')

# Display the figure
plt.show()
```

### **Question 6**: What is the impact of a new CRM system adoption on software-related incidents over time.

#### Impat of new CRM system adoption on software-related incidents

```python
import csv
from datetime import datetime
import matplotlib.pyplot as plt
from collections import defaultdict

def parse_date(date_string):
    return datetime.strptime(date_string, '%Y-%m-%d %H:%M:%S')

# Read and process the data
filename = 'csvs/flag-95.csv'
software_incidents = defaultdict(int)
crm_incidents = defaultdict(int)

with open(filename, 'r') as file:
    reader = csv.DictReader(file)
    for row in reader:
        date = parse_date(row['opened_at']).date()
        if row['category'] == 'Software':
            software_incidents[date] += 1
            if 'CRM' in row['short_description']:
                crm_incidents[date] += 1

# Prepare data for plotting
dates = sorted(software_incidents.keys())
software_counts = [software_incidents[date] for date in dates]
crm_counts = [crm_incidents[date] for date in dates]

# Create the plot
plt.figure(figsize=(12, 6))
plt.plot(dates, software_counts, label='All Software Incidents')
plt.plot(dates, crm_counts, label='CRM-related Incidents', color='red')

plt.title('Software Incidents Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Incidents')
plt.legend()

# Highlight the adoption period
adoption_start = datetime(2023, 7, 1).date()
adoption_end = datetime(2023, 9, 30).date()
plt.axvspan(adoption_start, adoption_end, color='yellow', alpha=0.3, label='CRM Adoption Period')

plt.xticks(rotation=45)
plt.tight_layout()

# Add annotations
max_crm_date = max(crm_incidents, key=crm_incidents.get)
plt.annotate(f'Peak CRM Issues: {crm_incidents[max_crm_date]}',
             xy=(max_crm_date, crm_incidents[max_crm_date]),
             xytext=(10, 10), textcoords='offset points',
             arrowprops=dict(arrowstyle="->"))

plt.legend()
plt.show()
plt.close()
```
